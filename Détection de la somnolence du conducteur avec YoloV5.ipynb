```python
# If not in ../ProjetComputerVisionESCRIBANO Go to ../ProjetComputerVisionESCRIBANO
%cd .. 
```

```python
SOURCE_PATH = os.path.join('SourcesData', 'SourcesForDatasetCreation') #/SourcesData/SourcesForDatasetCreation
IMAGES_PATH_Train = os.path.join('SourcesData', 'drowsinessDetection_datasetCreated', 'train', 'images') #/SourcesData/drowsinessDetection_datasetCreated/train
IMAGES_PATH_Valid = os.path.join('SourcesData', 'drowsinessDetection_datasetCreated', 'valid', 'images') #/SourcesData/drowsinessDetection_datasetCreated/valid
labels = ['awake', 'drowsy']
number_imgs = 50 # number of images per label (with 20% of it as validation set)
```

```python
# Loop through labels
for label in labels:
    path = os.path.join(SOURCE_PATH, label)
    files = os.listdir(path)
    i=0
    # loop through files
    for file in files:
        i+=1
        filepath = os.path.join(path, file)
        # Naming out image path
        if i <= number_imgs*0.8:
            imgname = os.path.join(IMAGES_PATH_Train, label+'.'+str(uuid.uuid1())+'.jpg')
        else:
            imgname = os.path.join(IMAGES_PATH_Valid, label+'.'+str(uuid.uuid1())+'.jpg')
            
        img = cv2.imread(filepath)
        
        # Writes out image to file 
        cv2.imwrite(imgname, img)
        if i == number_imgs:
            break
    print(label + ": Done")
        
```

### Real time detection roboflow dataset

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
